####Outliers Values

```python
# Mendefinisikan fungsi untuk mendeteksi outliers dan menghitung persentase
def detect_outliers(column):
    # Convert the column to a numeric data type
    column = pd.to_numeric(column, errors='coerce')

    # Remove NaN values
    column = column.dropna()

    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1

    # Detect outliers
    outliers = (column < (Q1 - 1.5 * IQR)) | (column > (Q3 + 1.5 * IQR))

    # Calculate the percentage of outliers
    percentage_outliers = outliers.sum() / len(column) * 100

    return outliers, percentage_outliers


# Loop untuk setiap kolom
for column in df.columns:
    outliers, percentage_outliers = detect_outliers(df[column])

    print(f"Kolom '{column}':")
    print(f"  - Jumlah outliers: {outliers.sum()}")
    print(f"  - Persentase kehadiran outliers: {percentage_outliers:.2f}%")
    print()
```

###**Penanganan Outliers**

#####**Pengecekkan Outliers**

```python

results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

results = []: Membuat sebuah list kosong yang akan digunakan untuk menyimpan hasil perhitungan persentase outlier untuk setiap kolom.

cols = df.select_dtypes(include=['float64', 'int64']): Memilih hanya kolom-kolom dalam DataFrame df yang memiliki tipe data numerik (float64 atau int64) dan menyimpannya dalam variabel cols.

Melakukan iterasi pada setiap kolom numerik yang dipilih dengan menggunakan loop for col in cols:.

Dalam setiap iterasi, dilakukan perhitungan terhadap outlier:

q1 = df[col].quantile(0.25): Menghitung kuartil pertama dari kolom tersebut.
q3 = df[col].quantile(0.75): Menghitung kuartil ketiga dari kolom tersebut.
iqr = q3 - q1: Menghitung rentang interkuartil (IQR) dari kolom tersebut.
lower_bound = q1 - 1.5*iqr: Menghitung batas bawah untuk outlier.
upper_bound = q3 + 1.5*iqr: Menghitung batas atas untuk outlier.
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]: Memilih baris-baris dalam DataFrame yang memiliki nilai di luar batas bawah atau batas atas yang dihitung sebelumnya.
percent_outliers = (len(outliers)/len(df))*100: Menghitung persentase outlier dari total jumlah baris dalam DataFrame df.
Menyimpan hasil perhitungan persentase outlier untuk setiap kolom dalam bentuk dictionary dan menambahkannya ke dalam list results.

Membuat DataFrame results_df dari list hasil dan melakukan pengaturan indeks serta nama kolom agar sesuai dengan kebutuhan.

#####**Hapus Outliers**

```python
Q1 = df['Year'].quantile(0.25)
Q3 = df['Year'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Year'] >= lower_bound) & (df['Year'] <= upper_bound)]
```

Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari kolom 'Year' menggunakan metode .quantile(). Kuartil pertama (Q1) adalah nilai yang membagi data menjadi 25% terbawah, sedangkan kuartil ketiga (Q3) adalah nilai yang membagi data menjadi 75% terbawah.

Menghitung rentang interkuartil (IQR) dengan mengurangkan Q1 dari Q3.

Menghitung batas bawah dan batas atas untuk outlier. Batas bawah dihitung dengan mengurangkan 1.5 kali IQR dari Q1, sedangkan batas atas dihitung dengan menambahkan 1.5 kali IQR dari Q3.

Membuat DataFrame baru df yang hanya berisi baris-baris di mana nilai 'Year' berada di antara batas bawah dan batas atas yang telah dihitung. Ini dilakukan dengan menggunakan pengindeksan boolean untuk memfilter baris-baris yang memenuhi kriteria tersebut

#####**Imputasi Outliers**

```python
Q1 = df['Global_Sales'].quantile(0.25)
Q3 = df['Global_Sales'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width * IQR)
upper_whisker = Q3 + (whisker_width * IQR)
df['Global_Sales']=np.where(df['Global_Sales']>upper_whisker,upper_whisker,np.where(df['Global_Sales']<lower_whisker,lower_whisker,df['Global_Sales']))
```

Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) dari kolom 'Global_Sales' menggunakan metode .quantile(). Kuartil pertama (Q1) adalah nilai yang membagi data menjadi 25% terbawah, sedangkan kuartil ketiga (Q3) adalah nilai yang membagi data menjadi 75% terbawah.

Menghitung rentang interkuartil (IQR) dengan mengurangkan Q1 dari Q3.

Menghitung whisker (tambahan dari batas bawah dan batas atas) untuk menentukan batas bawah dan batas atas untuk outlier. Batas bawah dihitung dengan mengurangkan whisker_width (dalam hal ini 1.5) kali IQR dari Q1, sedangkan batas atas dihitung dengan menambahkan whisker_width kali IQR dari Q3.

Menggunakan fungsi np.where() dari library NumPy untuk mengganti nilai 'Global_Sales' yang melebihi batas atas dengan nilai upper_whisker, dan nilai 'Global_Sales' yang kurang dari batas bawah dengan nilai lower_whisker. Nilai-nilai lainnya tetap tidak berubah

#####**Pengecekkan Ulang**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

#**Modelling Associate**

###**Pre-Processing**

#####**Adjusment**

```python
x_final = df.drop("NA_Sales", axis=1)
```

Perintah x_final = df.drop("NA_Sales", axis=1) digunakan untuk membuat variabel baru x_final yang berisi DataFrame dari df tanpa kolom "NA_Sales".

Jadi, x_final akan berisi DataFrame yang memiliki semua kolom yang sama seperti df, kecuali kolom "NA_Sales" yang telah dihapus. Dengan demikian, kolom "NA_Sales" tidak akan disertakan dalam x_final.

#####**Normalisasi Data**

```python
scaler = MinMaxScaler()

x_final_norm = scaler.fit_transform(x_final)
```

Perintah scaler = MinMaxScaler() membuat objek scaler yang akan menyesuaikan atau menormalkan data ke dalam rentang tertentu menggunakan Min-Max Scaling.

Kemudian, x_final_norm = scaler.fit_transform(x_final) menormalkan data dalam x_final menggunakan scaler yang telah dibuat sebelumnya. Data hasil normalisasi disimpan dalam x_final_norm.

#####**Validasi Silang**

```python
inertia_values = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_final_norm)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_range)
plt.show()
```

Berdasarkan grafik, nilai optimal untuk K adalah 4. Pada nilai ini, inersia mulai mencapai plateau, menunjukkan bahwa menambahkan lebih banyak klaster tidak akan memberikan banyak manfaat.

###**Build Model**

#####**K-Means**

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(x_final_norm)
```

Perintah kmeans = KMeans(n_clusters=4, random_state=42) digunakan untuk membuat objek kmeans dari kelas KMeans dengan 4 klaster yang akan digunakan untuk melakukan pengelompokan data.

Kemudian, kmeans.fit(x_final_norm) digunakan untuk melatih model KMeans pada data yang telah dinormalisasi (x_final_norm). Model KMeans akan menemukan pusat-pusat klaster yang optimal untuk membagi data ke dalam 4 klaster berdasarkan pola yang ada dalam data.

#####**Hierarchical Clustering**

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

Perintah hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward') digunakan untuk membuat objek clustering hierarkis aglomeratif (Agglomerative Clustering). Dalam pengaturan ini, ada empat klaster yang ditentukan (n_clusters=4). Metrik jarak yang digunakan adalah Euclidean (affinity='euclidean'). Selain itu, linkage method yang dipilih adalah Ward, yang merupakan salah satu metode penggabungan klaster dalam hierarki aglomeratif.

###**Begins and Compare Clusters**

```python
kmeans_clust = kmeans.predict(x_final_norm)
hierarchy_clust = hc.fit_predict(x_final_norm)
```

Perintah kmeans_clust = kmeans.predict(x_final_norm) digunakan untuk memprediksi klaster yang diberikan oleh model KMeans untuk setiap sampel dalam data yang telah dinormalisasi (x_final_norm). Hasil prediksi klaster disimpan dalam kmeans_clust.

Perintah hierarchy_clust = hc.fit_predict(x_final_norm) digunakan untuk melakukan clustering hierarkis aglomeratif pada data yang telah dinormalisasi (x_final_norm). Model clustering hierarkis aglomeratif (hc) dilatih dan kemudian digunakan untuk memprediksi klaster untuk setiap sampel dalam data. Hasil prediksi klaster disimpan dalam hierarchy_clust.

```python
x_final = pd.DataFrame(x_final).reset_index(drop=True)

kmeans_col = pd.DataFrame(kmeans_clust, columns=["kmeans_cluster"])
hierarchy_col = pd.DataFrame(hierarchy_clust, columns=["hierarchy_cluster"])

combined_data_assoc = pd.concat([x_final, kmeans_col, hierarchy_col], axis=1)
```

Perintah-perintah di atas digunakan untuk menggabungkan hasil clustering dari model KMeans dan model hierarchical clustering ke dalam DataFrame yang berisi data asli (x_final).
